```python
import os
import glob
import pandas as pd
import re

# Ruta a la carpeta que contiene los CSVs
data_folder = os.path.join("..","..", "data", "apertura_2025", "rounds")
data_folder_out = os.path.join("..","..", "data", "apertura_2025", "rounds", "out")

# Buscar todos los archivos que terminen en .csv
csv_files = glob.glob(os.path.join(data_folder, "*.csv"))

print(f"Archivos encontrados: {len(csv_files)}")
```

```python
dfs = []

for file in csv_files:
    try:
        df = pd.read_csv(file)

        # Nombre base del archivo
        base_name = os.path.basename(file).replace(".csv", "")

        # Obtener match_id: primeros 3 bloques separados por "_"
        parts = base_name.split("_")
        match_id = "_".join(parts[:3])

        # Obtener round_id usando regex (número después de "round_")
        match = re.search(r"round_(\d+)", base_name)
        round_id = int(match.group(1)) if match else None

        # Agregar columnas
        df['match_id'] = match_id
        df['round_id'] = round_id

        dfs.append(df)
    except Exception as e:
        print(f"Error leyendo {file}: {e}")

# Unir todos los dataframes
df_combined = pd.concat(dfs, ignore_index=True, sort=False)

# Guardar resultado final
output_path = os.path.join(data_folder_out, "rounds_unificado.csv")
df_combined.to_csv(output_path, index=False)

print(f"Se combinaron {len(csv_files)} archivos. Guardado en: {output_path}")
```
